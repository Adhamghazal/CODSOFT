```python
import random
```

```python
def cost_function(target,predict):
  #MSE
  m=len(target)
  SSE=0
  for i in range(len(target)):
    error=(target[i]-predict[i])**2
    SSE+=error
  return (SSE/(2*m)),SSE
```

```python
def gradient(features,target,iterations,LR):
  #Initialize thetas
  theta=[random.randint(1,10),random.randint(1,10)]
  mse=[]
  m=len(target)
  for itr in range(iterations):
    #predict and calculate error
    predict_y = predict(theta,features)
    cost,SSE=cost_function(target,predict_y)
    mse.append(cost)
    #update thetas
    theta[0] = theta[0] - (LR/m)*SSE
    temp_sse=0
    for i in range(len(target)):
      temp_sse += (predict_y[i]-target[i])* features[i]
    theta[1] = theta[1] - (LR/m)*temp_sse
  return theta,mse
```

```python
def predict(theta,features):
  #predicted values
  y_predict=[]
  for i in range(len(features)):
    y_predict.append(theta[0]+theta[1]* features[i])
  return y_predict
```

```python
x=[1,2,2,3,4,5,5]
y=[7,8,7,9,11,10,12]
theta,mse=gradient(x,y,10000,0.0001)
print(theta,mse[-1])
```
